```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = [
    ["Madinah", 2014, "Female", "Saudi", 562344],
]
```

```python
columns = ["Region", "Year", "Gender", "Nationality", "Population"]
df = pd.DataFrame(data, columns=columns)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x="Region", y="Population", estimator=sum)
plt.xticks(rotation=45)
plt.title("Total Population by Region")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x="Region", y="Population")
plt.xticks(rotation=45)
plt.title("Population Distribution by Region")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(data=df, x="Nationality", y="Population")
plt.title("Population Distribution by Nationality")
plt.show()
```

```python
sns.set(style="ticks")
sns.pairplot(df, hue="Nationality")
plt.title("Pairplot of the Data")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.swarmplot(data=df, x="Nationality", y="Population")
plt.title("Population Distribution by Nationality (Swarm Plot)")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Year", y="Population", hue="Nationality")
plt.title("Population vs. Year")
plt.show()
```

```python
gender_region = df.groupby(["Region", "Gender"])["Population"].sum().unstack()
gender_region.plot(kind="bar", stacked=True, figsize=(12, 6))
plt.title("Gender Distribution by Region (Stacked Bar Chart)")
plt.show()
```

```python
pivot_table = df.pivot_table(index="Region", columns="Year", values="Population")
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap="YlGnBu", annot=True, fmt=".1f")
plt.title("Population Heatmap by Region and Year")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x="Population", hue="Gender")
plt.title("Population Distribution by Gender")
plt.show()
```
